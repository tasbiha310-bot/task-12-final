```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("coretech_clients_cleaned.csv")

df.head()
```

```python
service_counts = df["service"].value_counts()

plt.figure(figsize=(8,5))
service_counts.plot(kind="bar")
plt.title("Clients by Service")
plt.xlabel("Service")
plt.ylabel("Number of Clients")
plt.savefig("bar_chart.png")
plt.show()
```

```python
lead_counts = df["lead_source"].value_counts()

plt.figure(figsize=(7,7))
plt.pie(
    lead_counts,
    labels=lead_counts.index,
    autopct="%1.1f%%"
)
plt.title("Lead Source Distribution")
plt.savefig("pie_chart.png")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(
    df["client_id"],
    df["project_budget"],
    marker="o"
)

plt.title("Project Budget Trend")
plt.xlabel("Client ID")
plt.ylabel("Project Budget")
plt.grid(True)
plt.savefig("line_chart.png")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(
    df["rating"],
    bins=5
)

plt.title("Ratings Distribution")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.savefig("histogram.png")
plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(
    df["project_duration_days"],
    df["project_budget"]
)

plt.title("Budget vs Duration")
plt.xlabel("Project Duration (Days)")
plt.ylabel("Project Budget")
plt.savefig("scatter_plot.png")
plt.show()
```
